```julia
# tworzenie macierzy
function createAdjMatrix(V, G)
    adjMatrix = []

    # tworzenie macierzy z 0 wartością wagi
    for i in range(0, V)
        adjMatrix.append([])
        for j in range(0, V)
            adjMatrix[i].append(0)

    # wypęłnienie macierzy odpowiednimi wagami
    for i in range(0, len(G))
        adjMatrix[G[i][0]][G[i][1]] = G[i][2]
        adjMatrix[G[i][1]][G[i][0]] = G[i][2]

    return adjMatrix


function prims(V, G)
    #tworzenie macierzy na podstawie grafu
    adjMatrix = createAdjMatrix(V, G)

    #narzucenie wyboru wierzchołku 
    vertex = 0

    #inicjacja pustej macierzy krawędzi i puste MST
    MST = []
    edges = []
    visited = []
    minEdge = [None, None, float('inf')]

    #pracoować tak długo aż wypełni wszystkie puste miejsca
    while len(MST) != V - 1:

        #zaznaczenie odwiedzenia wierzchołka
        visited.append(vertex)

        # dodanie każdej krawędzi do listy potencjalnych krawędzi 
        for r in range(0, V):
            if adjMatrix[vertex][r] != 0:
                edges.append([vertex, r, adjMatrix[vertex][r]])

        # znajdź krawędź najmniejszej wartości, której jeszcze nie użyto
        for e in range(0, len(edges)):
            if edges[e][2] < minEdge[2] and edges[e][1] not in visited:
                minEdge = edges[e]

       #usunięcie minimalnej wagi krawędzi
        edges.remove(minEdge)

        # wrzucenie minimalnej wagi do MST
        MST.append(minEdge)

        #zacznij od nowego wierzchołka i zresetuj minimalne wagi
        vertex = minEdge[1]
        minEdge = [None, None, float('inf')]

    return MST


a, b, c, d, e, f = 0, 1, 2, 3, 4, 5

# nasz graf
graph = [
    [a, b, 2],
    [a, c, 3],
    [b, d, 3],
    [b, c, 5],
    [b, e, 4],
    [c, e, 4],
    [d, e, 2],
    [d, f, 3],
    [e, f, 5]
]



print(prims(6, graph))

```
